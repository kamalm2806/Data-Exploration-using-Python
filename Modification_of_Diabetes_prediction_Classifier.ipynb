```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
dataset = convert(ds)
dataset.head(100)
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
```

# Pairgrid Graph
An thorough comparison study is made possible by using the PairGrid graph to represent pairwise correlations among multiple classes' worth of attributes. Possible connections or trends between various class-specific features may be seen using this graphic.

```python

g = sns.PairGrid(du)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.suptitle('Distribution of classes')
plt.show()
```

# KDE Plot

One way to visualize data is via a KDE (Kernel Density Estimation) plot, which takes a continuous variable and uses it to estimate its probability density function. A smooth curve is used to depict the data distribution. If you want to see how your data is distributed and how the probability density changes for various values, a KDE plot is a good tool to use.


```python
sns.kdeplot(du['BMI'], shade=True, color='orange')
plt.title('BMI vs DP Function')
plt.xlabel('BMI')
plt.ylabel('DP FUnction')
plt.show()
```

# Jointplot Graph

The link between two characteristics in separate classes may be shown visually in the Joint plot graph. The correlations and interactions between these traits across different classes are highlighted, providing useful insights into their relationships.




```python

sns.jointplot(x='DP Function', y='PG Concentration', data=du, kind='scatter')
plt.title('Joint plot of DP Function vs. PGConcentration')
plt.show()

```

# Regression Plot

One way to visualize data is using a regression plot, which shows the linear connection between two variables. This kind of plot is often made with a linear regression line. You may learn a lot about the direction and strength of a relationship—and how one variable changes in respect to another—by looking at this graphic.


```python

plt.figure(figsize=(8, 6))
sns.regplot(x='Pregnancies', y='BMI', data=du)
plt.title('Regression plot of Pregnancies vs BMI')
plt.xlabel('Pregnancies')
plt.ylabel('BMI')
plt.show()
```

# Histogram

One way to visually depict the distribution of numerical data is via a histogram. It shows how often values fall into certain intervals or bins. Learn more about the central tendency, dispersion, and skewness of a continuous variable's data, as well as its underlying frequency distribution, with the use of a histogram.


```python

# Histogram
sns.histplot(data=du, x='Age', bins=20, color='Orange', alpha=0.9, label='Age')
sns.histplot(data=du, x='BMI', bins=20, color='Pink', alpha=0.9, label='BMI')
plt.title('Histogram of Age and BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend()
plt.show()



```

# Swarnplot Graph

By presenting individual data points, the Swarm plot graph illustrates how a characteristic is distributed across distinct classes. The distribution and concentration of data for each class may be better seen with its support.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Outcome' is the target variable (0 for non-diabetic, 1 for diabetic)
plt.figure(figsize=(50, 20))
sns.violinplot(x='Tri Fold Thick', y='BMI', data=du, inner=None, color='lightgray')
sns.swarmplot(x='Tri Fold Thick', y='BMI', data=du, color='purple', alpha=0.7)
plt.title('Diabetes by Tri Fold Thick vs BMI')
plt.xlabel('Tri_Fold_Thick (0: Non-diabetic, 1: Diabetic)')
plt.ylabel('BMI')
plt.show()
```
